```python
df = pd.read_csv(f"{FILEPATH}/states_all.csv")
df.columns
```

```python
columns_to_keep = [
    "STATE",
    "YEAR",
    "TOTAL_REVENUE",
    "TOTAL_EXPENDITURE",
    "GRADES_8_G",
    "AVG_MATH_8_SCORE",
    "AVG_READING_8_SCORE",
]

df = df[columns_to_keep]
```

```python
df["AVG_MATH_8_SCORE"].describe()
```

```python
df['DIFFERENCE'] = df['TOTAL_REVENUE'] - df['TOTAL_EXPENDITURE']
df['EXPENDITURE_PER_STUDENT'] = df['TOTAL_EXPENDITURE'] / df['GRADES_8_G']


df['REVENUE_PER_STUDENT'] = df['TOTAL_REVENUE'] / df['GRADES_8_G']
df['REVENUE_EXPENDITURE_RATIO'] = df['TOTAL_REVENUE'] / df['TOTAL_EXPENDITURE']
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

df.fillna(df.median(), inplace=True)

# Assuming you have a dataframe df
# 1. Identify numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.drop('YEAR')

# 2. Apply StandardScaler to only numerical columns
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```
